```python
def str_to_list(x: Optional[str]=None):
    return [x] if x and isinstance(x, str) else x

def join_conds(conds: Optional[dict]=None):
    return " and ".join(f"{k} == '{v}'" for k, v in conds.items())

def make_grouping(
        filter: dict=None,
        all_same: Optional[Union[str, Sequence]]=None,
        all_diff: Optional[Union[str, Sequence]]=None,
        any_same: Optional[Union[str, Sequence]]=None,
        any_diff: Optional[Union[str, Sequence]]=None):
    return {
        "filter": filter,
        "all_same": str_to_list(all_same),
        "all_diff": str_to_list(all_diff),
        "any_same": str_to_list(any_same),
        "any_diff": str_to_list(any_diff)
    }

tech_rep_config = {
    # "aggregate": None,
    "group": make_grouping(
        filter={"Metadata_pert_type": "trt"},
        all_same="Metadata_JCP2022",
        any_diff=["Metadata_Plate", "Metadata_Well"]),
    "reference": make_grouping(
        filter={"Metadata_pert_type": "negcon"},
        any_diff=["Metadata_Plate", "Metadata_Well"])
}

bio_rep_config = {
    # "aggregate": None,
    "group": make_grouping(
    # alternative to filtering allow adding a condition to the groupby
        all_same=["Metadata_JCP2022", "Metadata_pert_type=='trt'"]
    ),
    "pair": {
        # each to each
        # each to one (define which)
        # each to some (define which)
    }
}
```
